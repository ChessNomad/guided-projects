```python
%%sql
SELECT MIN(population),MAX(population), MIN(population_growth), MAX(population_growth)
  FROM facts
 WHERE Population < 7000000000
```

```python
%%sql
SELECT AVG(population), AVG(area)
  FROM facts
 WHERE population < 7000000000
```

```python
%%sql
SELECT name, population, area
  FROM facts
 WHERE population > (SELECT AVG(population)
                       FROM facts
                      WHERE population < 7000000000)
   AND area < (SELECT AVG(area)
                 FROM facts
                WHERE population < 7000000000);
```

```python
%%sql
SELECT MAX(population), name
  FROM facts
 WHERE Population < 7000000000
```

```python
%%sql
SELECT name, MAX(population_growth)
  FROM facts
 WHERE Population < 7000000000;
```

```python
%%sql 
SELECT name, (CAST(area_water AS float) / area_land) AS ratio
  FROM facts
 WHERE Population < 7000000000
 ORDER BY ratio DESC
 LIMIT 10;
```

```python
%%sql 
SELECT name
  FROM facts
 WHERE area_water > area_land
   AND Population < 7000000000
```

```python
%%sql 
SELECT name, population * population_growth / 100 AS people_add
  FROM facts
 WHERE population < 7000000000
 ORDER BY people_add DESC
 LIMIT 20;
```

```python
%%sql 
SELECT name
  FROM facts
 WHERE population < 7000000000
   AND death_rate > birth_rate
```

```python
%%sql
SELECT name, population/area AS ratio
  FROM facts
 WHERE population < 7000000000
 ORDER BY ratio DESC;
```
