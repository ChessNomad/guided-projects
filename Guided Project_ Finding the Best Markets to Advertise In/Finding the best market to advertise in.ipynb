# E-Learning marketing strategy
Most of our courses are on web and mobile development, but we also cover many other domains, like data science, game development, etc. We want to promote our product and we'd like to invest some money in advertisement.

Our goal in this project is to find out the two best markets to advertise our product in.

To reach our goal, we could organize surveys for a couple of different markets to find out which would the best choices for advertising. This is very costly, however, and it's a good call to explore cheaper options first. One good candidate is the data from freeCodeCamp's 2017 New Coder [Survey](https://www.freecodecamp.org/news/we-asked-20-000-people-who-they-are-and-how-theyre-learning-to-code-fff5d668969/). freeCodeCamp is a free e-learning platform that offers courses on web development. Because they run a popular Medium [publication](https://www.freecodecamp.org/news/) (over 400,000 followers), their survey attracted new coders with varying interests (not only web development), which is ideal for the purpose of our analysis.

As we can see the top 5 has changed considerably after we have exploded the data. However there is one little problem. ### A new focus
Our company is mostly interested in 'web' and 'mobile' developers. Therefore it is wise to add a new column to the original data, which tells us the percentage of web/mobile vs. non-web/mobile. 

```python
#let's search the original dataset for occurances of 'web' and 'mobile' with regular expressions.
#First we need to get rid of the missing values, value_counts() will do that for us.
web_mobile = pd.DataFrame(code_survey['JobRoleInterest'].dropna())

#Now we can search the sample.
web_mobile['?'] = web_mobile['JobRoleInterest'].str.contains('[wW]eb,?|[Mm]obile,?')
web_mobile['?'].value_counts(normalize=True)*100
```

### Web or Mobile

# Location, location, location
Because most people in our population fit our target group, we would like to move on to the next part of our marketing strategy. Which is as you may have guessed, location. 

Next we would like to identify the best markets for our advertisements by size. The more potential customers the better. In order to find the best two markets we will use the CountryLive column, because it is more important where people live then where they are from (CountryFrom is another column). 

One of the discussions here is if the NaN values can be used. Well let's first see if the answers are very different in order to consider what to do.

```python
pd.set_option('display.max_rows', None)
#Firstly we will drop the null values.
job_interest = code_survey[code_survey['JobRoleInterest'].notnull()]

#And count the values in absolutes and percentages.
print(job_interest['CountryLive'].value_counts())
job_interest['CountryLive'].value_counts(normalize=True)*100
```

```python
pd.set_option('display.max_rows', None)

#Now for the full dataset.
code_survey['CountryLive'].value_counts(normalize=True)*100
```

## Two winners, four contestants
Fortunately we have found that with or without the null values the two biggest markets to advertise in are the USA and India.

However we should still continue our analysis to see if the markets have a willingness to spend money. A country where most people like to study for free is not going to yield a lot of results, no matter how big the interest is. Therefore it would be good to determine the total amount of money they are willing to spend.

For that we will narrow our search down to the top four countries in order to see where the budget is the biggest. The definition of money to spend will be money spent per month. Which means we will select only the rows from the dataset where the country is USA, India, UK or Canada. Another reason for picking these countries is that English is an official language in these countries. 

```python
#First we will extract the country column from the null-filtered dataset.
job_interest = job_interest[job_interest['CountryLive'].isin(['United States of America', 'India', 'United Kingdom','Canada'])]
job_interest['CountryLive'].value_counts(dropna=False)
#We can see there are no null values, and only the countries we selected. 
```

```python
#In order to get the budget per month we need to divide MoneyForLearning by MonthsProgramming. 
#For MonthsProgramming the value can't be 0, so we need to change that to 1 first.
job_interest["MonthsProgramming"] = job_interest['MonthsProgramming'].replace(0,1)

#Now we are safe from dividing by 0.
job_interest["MonthlyBudget"] = (job_interest['MoneyForLearning'] / job_interest['MonthsProgramming'])
job_interest = job_interest[job_interest["MonthlyBudget"].notnull()]
job_interest['MonthlyBudget'].head()
```

```python
#Creating a table for the mean prices per country.
table = pd.pivot_table(job_interest, values='MonthlyBudget', index='CountryLive')
table
```

Intuitively these results seem a bit strange because the [GDP per capita](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita) indicates otherwise. There are two explanations, either there are a lot of outliers for UK or India or the results are correct and that means that only richer people from India pick up coding. Let's check this.

```python
#We will use a boxplot to check for outliers.
job_interest.boxplot(column='MonthlyBudget', by='CountryLive', figsize=(10,7))
plt.show()
```

As we can see there are some extreme outliers in the USA. Let us check what the people who spent above 5000$ spent their money on manually. 

```python
#adjusting settings and selecting the highest values for inspection
pd.set_option('display.max_rows', None)
job_interest_5k = job_interest[job_interest['MonthlyBudget'] >= 5000]
print(job_interest_5k)
```

We can see that the people from India who spent more than \\$5000 did not do a bootcamp and neither did the major outliers from the US. Therefore I will drop the people who have a monthly budget of more than \\$5000 and did not attend a bootcamp.

```python
#We filter for under 5000 or (over and attended bootcamp).
job_interest = job_interest[(job_interest['MonthlyBudget'] < 5000) | ((job_interest['MonthlyBudget'] >= 5000) & job_interest['AttendedBootcamp']==1.0)]
job_interest['MonthlyBudget'].max()
```

```python
#Still it is a bit strange to spend over 20000 per month, so let's remove that anyway.
job_interest = job_interest[(job_interest['MonthlyBudget'] < 20000)]

#Now we can do a similar boxplot to above
job_interest.boxplot(column='MonthlyBudget', by='CountryLive', figsize=(10,7))
plt.show()
```

```python
#Maybe the values have changed now when we have removed the biggest outliers.
table2 = pd.pivot_table(job_interest, values='MonthlyBudget', index='CountryLive')
table2
```

## USA
As we can see USA is a clear target for our ad. New coders are willing to spend a lot however the second country is still undecided. Canada seems to be the clear winner, however we have to take into account that our price is \\$57 per course. Which a lot of people in India could also afford. 

Let's see what percentage of individuals can afford \\$57 or more before we decide on our next step. I will take \\$50 as a benchmark because these people might be willing to spend a little more than they think if the like the product. 

```python
#First create a new column
job_interest['AffordCourse'] = job_interest['MonthlyBudget'] >= 50 

#Then create values for afford or not
table3 = pd.pivot_table(job_interest, index=['CountryLive','AffordCourse'],aggfunc='size')
table3
```

```python
#Let's quickly divide the result manually
print('Canada', 45/195)
print('India',78/381)
print('UK',49/230)
print('USA', 737/2187)
```

# Divided results
It is clear that 34% in the USA is able to afford our course. Combined with the amount of people USA is a clear winner. Second is Canada with only 23%. This difference does not seem too big, however, compared to USA where 1 in 3 are able to afford the course it is quite big to see 1 in 5. This could mean losing 11% of the targeted customers. 

All in all the end decision is for our marketing department, however I would like to recommend to choose the USA only both by absolute size of the population and affordability. If after a while it seems we are not getting as many customers we could still try India as a market. The population is larger than the others, and the percentage of people who are able to pay is not much less.
