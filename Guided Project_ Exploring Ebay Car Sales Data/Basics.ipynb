### Numerics
Let us investigate the numeric aspect of the cleaning:
* nr_of_pictures has values of 0.0 only, hence that column is not interesting and will be deleted. 
* seller has the same values and will be deleted as well.
* The max year is 9999 which is a bit odd.
* The horsepower(ps) max value is also skyhigh, I believe there had to be a rocket for sale ;)

```python
# Let us correct the price and odometer columns and convert it to int.
autos["price"] = autos["price"].str.replace("$","").str.replace(",","")
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","")

# now converting them to integers
autos["price"] = autos["price"].astype(int)
autos["odometer"] = autos["odometer"].astype(int)

autos.rename(columns={"odometer":"odometer_km"}, inplace = True)
autos.head()
```

## Brands mean price 
In order to have a large enough dataset for an accurate calculation of the mean price, I will select the brands that have at least 1% in the dataset. 500 entries should be enough to create a reliable value for mean price per brand.

```python
autos["brand"].value_counts(normalize = 1).index
```

```python
# The brands that have more than 1% are up to hyundai
brands1pct = ['volkswagen', 'bmw', 'opel', 'mercedes_benz', 'audi', 'ford', 'renault',
       'peugeot', 'fiat', 'seat', 'skoda', 'nissan', 'mazda', 'smart',
       'citroen', 'toyota', 'hyundai']
```

```python
# let's aggegrate and create a dictionary which shows mean price
# for the brands1pct.
brands1pct_mean_price = {}

for b in brands1pct:
    brand = autos[autos["brand"] == b]
    mean_price = brand["price"].mean()
    mean_price = mean_price.round()
    brands1pct_mean_price[b] = mean_price
    
print(brands1pct_mean_price)
```

```python
# Right now we will make another dictionary in order to compare the
# mean price to the mean mileage.
brands1pct_mean_mileage = {}

for b in brands1pct:
    brand = autos[autos["brand"] == b]
    mean_mileage = brand["odometer_km"].mean()
    mean_mileage = mean_mileage.round()
    brands1pct_mean_mileage[b] = mean_mileage
    
print(brands1pct_mean_mileage)
```
